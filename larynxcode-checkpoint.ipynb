```python
def correlation_based_feature_selection(X, correlation_threshold = 0.8):
    correlation_matrix = X.corr()
    
    features_to_remove = set()
    
    for i in range(len(correlation_matrix.columns)):
        for j in range(i + 1, len(correlation_matrix.columns)):
            if abs(correlation_matrix.iloc[i, j]) >= correlation_threshold:
                feature_i = correlation_matrix.columns[i]
                feature_j = correlation_matrix.columns[j]

            
                if feature_i not in features_to_remove:
                    features_to_remove.add(feature_i)
    return features_to_remove

def calculate_accuracy(y_true, y_pred):
    return np.sum(y_true == y_pred) / y_pred.shape[0]
```

```python
radiomics_data = pd.read_csv('LNmet.csv')
X = radiomics_data.drop(columns=['Response'])
y = radiomics_data['Response']
X.shape, y.shape
```

# Cross Validation

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.base import clone
import numpy as np
import pandas as pd


start_column_index = 0
end_column_index = 117  
normalize_columns = True  

scaler = MinMaxScaler()

test_folds_list = []
trained_model_folds_list = []

tprs = []
aucs = []
mean_fpr = np.linspace(0, 1, 100)

stratified_kfold = StratifiedKFold(n_splits=number_of_fold)
trn_idx, tst_idx = next(iter(stratified_kfold.split(X, y)))

for cv_idx, (trn_idx, tst_idx) in enumerate(stratified_kfold.split(X, y)):
    classifier_ = clone(classifier)
    X_train, X_test, y_train, y_test = X.iloc[trn_idx],X.iloc[tst_idx],y.iloc[trn_idx],y.iloc[tst_idx]
    
    if normalize_columns:
        
        columns_to_normalize = X_train.columns[start_column_index:end_column_index]
        X_train[columns_to_normalize] = scaler.fit_transform(X_train[columns_to_normalize])
        X_test[columns_to_normalize] = scaler.transform(X_test[columns_to_normalize])

    
    print("\nCross Validation Index: {}".format(cv_idx))
    print("Array shapes", X_train.shape, X_test.shape, y_train.shape, y_test.shape)

    features_to_remove = correlation_based_feature_selection(X_train)

    if eliminate_features_by_corr:
        
        X_train_filtered = X_train.drop(columns=features_to_remove)
        X_test_filtered = X_test.drop(columns=features_to_remove)
    else:
        X_train_filtered = X_train
        X_test_filtered = X_test

    if eliminate_features_select_k:
        selector = SelectKBest(score_func=f_regression, k=5)
        selector.fit(X_train_filtered, y_train)
        
        top_5_feature_indices = selector.get_support(indices=True)
        
        top_5_features = X_train_filtered.columns[top_5_feature_indices]
        
        X_train_filtered = X_train_filtered[top_5_features]
        X_test_filtered = X_test_filtered[top_5_features]
    else:
        X_train_filtered = X_train
        X_test_filtered = X_test
        
    print("Array shapes after filtering", X_train_filtered.shape, X_test_filtered.shape, y_train.shape, y_test.shape)
    print("\n")
    if eliminate_features_select_k:
        print("Cross Validation Idx : {}, top-k features : {}\n".format(cv_idx, top_5_features))
        
    classifier_.fit(X_train_filtered, y_train)

    y_train_pred = classifier_.predict(X_train_filtered)
    # y_train_proba = classifier_.predict_proba(X_train)

    y_test_pred = classifier_.predict(X_test_filtered)
    # y_test_pred_proba = classifier_.predict_proba(X_test)
    
    print("train accuracy: ", calculate_accuracy(y_train.values, y_train_pred), "test accuracy: ",  calculate_accuracy(y_test.values, y_test_pred))

    cm = confusion_matrix(y_test, y_test_pred)

    ConfusionMatrixDisplay(confusion_matrix=cm,
                                display_labels=classifier_.classes_).plot()
    
    test_folds_list.append((X_test_filtered, y_test))
    trained_model_folds_list.append(classifier_)
    print("################################################## \n")
fig, ax = plt.subplots(figsize=(6, 6))
for i in range(len(test_folds_list)):
    X_test_filtered, y_test = test_folds_list[i]
    classifier_ = trained_model_folds_list[i]
    viz = RocCurveDisplay.from_estimator(   classifier_, X_test_filtered, y_test, pos_label = 1,
                                            name=f"ROC fold {i}",
                                            alpha=0.3,
                                            lw=1,
                                            ax=ax,)
    interp_tpr = np.interp(mean_fpr, viz.fpr, viz.tpr)
    interp_tpr[0] = 0.0
    tprs.append(interp_tpr)
    aucs.append(viz.roc_auc)
    
    print("\nFalse positive rates and True positive rates for each fold\n", viz.fpr, viz.tpr)
```

# Plot Averaged ROC  

```python
fig, ax = plt.subplots(figsize=(6, 6))
mean_tpr = np.mean(tprs, axis=0)
mean_tpr[-1] = 1.0
mean_auc = auc(mean_fpr, mean_tpr)
std_auc = np.std(aucs)
ax.plot(
    mean_fpr,
    mean_tpr,
    color="b",
    label=r"Mean ROC (AUC = %0.2f $\pm$ %0.2f)" % (mean_auc, std_auc),
    lw=2,
    alpha=0.8,
)

std_tpr = np.std(tprs, axis=0)
tprs_upper = np.minimum(mean_tpr + std_tpr, 1)
tprs_lower = np.maximum(mean_tpr - std_tpr, 0)
ax.fill_between(
    mean_fpr,
    tprs_lower,
    tprs_upper,
    color="grey",
    alpha=0.2,
    label=r"$\pm$ 1 std. dev.",
)

ax.set(
    xlim=[-0.05, 1.05],
    ylim=[-0.05, 1.05],
    xlabel="False Positive Rate",
    ylabel="True Positive Rate",
    title=f"Mean ROC curve with variability\n(Positive label)",
)
ax.axis("square")
ax.legend(loc="lower right")
plt.show()
```

```python
import numpy as np
from sklearn.metrics import roc_curve


def find_best_threshold_youden(y_true, y_pred_proba):
    fpr, tpr, thresholds = roc_curve(y_true, y_pred_proba)
    youden_index = tpr - fpr
    best_threshold_index = np.argmax(youden_index)
    best_threshold_youden = thresholds[best_threshold_index]
    return best_threshold_youden, youden_index[best_threshold_index]


best_threshold_test_youden, best_youden_index_test = find_best_threshold_youden(y_test.values, y_test_pred_proba)

print("Best threshold and Youden index on test data:", best_threshold_test_youden, best_youden_index_test)
```
